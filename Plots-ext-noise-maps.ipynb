# Extrinsic noise plots for paper

```julia
# import all the relevant packages.
using Plots, LaTeXStrings, Distributions, DoubleFloats, Plots.PlotMeasures, LinearAlgebra, DataFrames, CSV
Plots.theme(:dao)
```

```julia
pal = cgrad(:tol_muted);
cpal = [pal[i] for i in 0.1:0.1:1.0]
```

# Extrinsic noise on $\rho_b$

```julia
ρb_hm_data = CSV.read("./ext-plots/rho_b_hm.csv", DataFrame, header = false);

hm1 = scatter(ρb_hm_data[:,1], ρb_hm_data[:,2], zcolor=ρb_hm_data[:,3], marker=:circle, markersize=20, 
        color=:roma, legend=false, colorbar=true, xaxis = :log, yaxis = :log, size = (350,300), 
            yticks = [10,10^1.5,10^2], markershape = :square, right_margin = 10mm)
plot!(xlabel = L"\rho_{\mathrm{off}}", ylabel = L"\rho_{\mathrm{on}}", xlims = [0.01,1], ylims = [10,100])
annotate!(5,10^1.5,L"e_{\rho_{\mathrm{on}}}")
plot!(title = "(b)", titleloc = :left)
# savefig("./ext-plots/rho_b_heat_map.svg")
```

Now for the line plots.

```julia
ρb_line1 = CSV.read("./ext-plots/rho_b_p1.csv", DataFrame, header = false);
ρb_line2 = CSV.read("./ext-plots/rho_b_p2.csv", DataFrame, header = false);
ρb_line3 = CSV.read("./ext-plots/rho_b_p3.csv", DataFrame, header = false);
```

```julia
lp1 = plot(ρb_line1[:,1],ρb_line1[:,2], xaxis = :log, c = cpal[1], label = L"\rho_{\mathrm{off}} = 1", lw = 3)
plot!(ρb_line2[:,1],ρb_line2[:,2], xaxis = :log, c = cpal[3], label = L"\rho_{\mathrm{off}} = 0.1", lw = 3)
plot!(ρb_line3[:,1],ρb_line3[:,2], xaxis = :log, c = cpal[8], label = L"\rho_{\mathrm{off}} = 0.01", lw = 3)
plot!(grid = false, minorgrid = false, legend = :topleft, size = (200,300), 
        xlabel = L"Var$\,(\rho_{\mathrm{on}})/\langle \rho_{\mathrm{on}}\rangle$", ylabel = L"e_{\rho_{\mathrm{on}}}")
plot!(title = "(a)", titleloc = :left)

savefig("./ext-plots/rho_b_lines.svg")
```

# Extrinsic noise on $\sigma_u$

```julia
σu_hm_data = CSV.read("./ext-plots/sigma_u_hm.csv", DataFrame, header = false);

hm2 = scatter(σu_hm_data[:,1], σu_hm_data[:,2], zcolor=σu_hm_data[:,3], marker=:circle, markersize=20, 
        color=:roma, legend=false, colorbar=true, xaxis = :log, yaxis = :log, size = (350,300), 
        markershape = :square, clims = (-0.55,0.1))
plot!(xlabel = L"e", ylabel = L"\langle\sigma_{\mathrm{off}}\rangle", 
xlims = [0.1,10],ylims = [0.01,100])
plot!([0.1,10],[1,1], color = :black)
annotate!(50,1,L"e_{\sigma_{\mathrm{off}}}")
annotate!(1,0.55,L"$\langle \sigma_{\mathrm{off}}\rangle=\sigma_{\mathrm{on}}$")
plot!(title = "(d)", titleloc = :left, right_margin = 12mm)

savefig("./ext-plots/sigma_u_heat_map.svg")
```

Now for the line plots.

```julia
σu_line1 = CSV.read("./ext-plots/sigma_u_p1.csv", DataFrame, header = false);
σu_line2 = CSV.read("./ext-plots/sigma_u_p2.csv", DataFrame, header = false);
σu_line3 = CSV.read("./ext-plots/sigma_u_p3.csv", DataFrame, header = false);
```

```julia
lp2 = plot(σu_line1[:,1],σu_line1[:,2], xaxis = :log, c = cpal[1], label = L"e = 0.1", lw = 3)
plot!(σu_line2[:,1],σu_line2[:,2], xaxis = :log, c = cpal[3], label = L"e = 1", lw = 3)
plot!(σu_line3[:,1],σu_line3[:,2], xaxis = :log, c = cpal[8], label = L"e = 10", lw = 3)
plot!([0.1,10],[0,0],xlims = [0.1,10], color = :black, label = :none)
plot!(grid = false, minorgrid = false, legend = :bottomleft, size = (200,300), 
        xlabel = L"Var$\,(\sigma_{\mathrm{off}})/\langle \sigma_{\mathrm{off}}\rangle$", ylabel = L"e_{\sigma_{\mathrm{off}}}")
plot!(title = "(c)", titleloc = :left)

savefig("./ext-plots/sigma_u_lines.svg")
```

# Extrinsic noise on $d$

```julia
d_hm_data = CSV.read("./ext-plots/d_hm.csv", DataFrame, header = false);

hm3 = scatter(d_hm_data[:,1], d_hm_data[:,2], zcolor=d_hm_data[:,3], marker=:circle, markersize=20, 
        color=:roma, legend=false, colorbar=true, colorbar_ticks = (0.2:0.2:1, string.(0.2:0.2:1)), 
            xaxis = :log, yaxis = :log, size = (350,300))
plot!(xlabel = L"\sigma_{\mathrm{off}}", ylabel = L"\langle e\rangle", right_margin =10mm, xlims=[0.01,100], ylims=[0.1,10])
annotate!(2500,1,L"e_e")
plot!(title = "(f)", titleloc = :left)

savefig("./ext-plots/d_heat_map.svg")
```

Now for the line plots.

```julia
d_line1 = CSV.read("./ext-plots/d_p1.csv", DataFrame, header = false);
d_line2 = CSV.read("./ext-plots/d_p2.csv", DataFrame, header = false);
d_line3 = CSV.read("./ext-plots/d_p3.csv", DataFrame, header = false);
```

```julia
lp3 = plot(d_line1[:,1],d_line1[:,2], xaxis = :log, c = cpal[1], label = L"\sigma_{\mathrm{off}} = 1", lw = 3)
plot!(d_line2[:,1],d_line2[:,2], xaxis = :log, c = cpal[3], label = L"\sigma_{\mathrm{off}} = 10", lw = 3)
plot!(d_line3[:,1],d_line3[:,2], xaxis = :log, c = cpal[8], label = L"\sigma_{\mathrm{off}} = 100", lw = 3)
plot!(grid = false, minorgrid = false, legend = :topleft, size = (200,300), 
        xlabel = L"Var$\,(e)/\langle e\rangle$", ylabel = L"e_{e}")
plot!(title = "(e)", titleloc = :left)

savefig("./ext-plots/d_lines.svg")
```

```julia
layout = @layout([a{0.4w} b{0.6w}; c{0.4w} d{0.6w}; e{0.4w} f{0.6w}])
plot(lp1, hm1, lp2, hm2, lp3, hm3, layout = layout, size = (700,900), left_margin = 5mm)
# savefig("./ext-plots/comb-ext-noise-plot.svg")
```
